```python
TCA = FR_BS.columns.get_loc("Total Current Assets")
TNCA = FR_BS.columns.get_loc("Total Noncurrent Assets")
TCL = FR_BS.columns.get_loc("Total Current Liabilities")
TNCL = FR_BS.columns.get_loc("Total Noncurrent Liabilities")
TE = FR_BS.columns.get_loc("Total Equity")

FR_BS["_Total Current Assets"] = FR_BS.iloc[:,:TCA].sum(axis=1)
FR_BS["_Total Noncurrent Assets"] = FR_BS.iloc[:,TCA+1:TNCA].sum(axis=1)
FR_BS["_Total Current Liabilities"] = FR_BS.iloc[:,TNCA+2:TCL].sum(axis=1)
FR_BS["_Total Noncurrent Liabilities"] = FR_BS.iloc[:,TCL+1:TNCL].sum(axis=1)
FR_BS["_Total Equity"] = FR_BS.iloc[:,TNCL+2:TE-1].sum(axis=1)
```

```python
import pandas as pd
import numpy as np

BS_Check = pd.DataFrame()
BS_Check["Total Assets"] = FR_BS["_Total Current Assets"] + FR_BS["_Total Noncurrent Assets"]
BS_Check["Total Liabilities"] = FR_BS["_Total Current Liabilities"] + FR_BS["_Total Noncurrent Liabilities"]
BS_Check["Total Equity"] = FR_BS["_Total Equity"]
BS_Check["Total Liabilities & Equity"] = BS_Check["Total Liabilities"] + BS_Check["Total Equity"]
BS_Check["Check"] = np.where(BS_Check["Total Assets"] == BS_Check["Total Liabilities & Equity"], 'True', 'False')
```
